```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
N = 50

X1 = np.random.randn(N, 2) + np.array([2, 2])  # Class 1 centered at (2,2)
X2 = np.random.randn(N, 2) + np.array([-2, -2])  # Class 2 centered at (-2,-2)

# Merge data
X = np.vstack((X1, X2))
y = np.hstack((np.ones(N), -np.ones(N)))  # Labels: 1 for Class 1, -1 for Class 2

# Plot data
plt.scatter(X1[:,0], X1[:,1], color='blue', label='Class 1 (+1)')
plt.scatter(X2[:,0], X2[:,1], color='red', label='Class 2 (-1)')
plt.legend()
plt.show()
```

```python
# Initialize weight vector (a) and bias (b)
a = np.random.randn(2)
b = np.random.randn()
```

```python
def train_perceptron(x, y, lr=0.1, epochs=200):
    global a, b
    for epoch in range(epochs):
        for i in range(len(x)):
            if y[i] * (np.dot(a, X[i]) + b) <= 0:  # Misclassification
                a += lr * y[i] * X[i]  # Update weights
                b += lr * y[i]  # Update bias
```

```python
train_perceptron(x, y)
```

```python
def predict(x):
    return np.sign(np.dot(a, x) + b)
```

```python
print(predict([0, 0])) 
```
